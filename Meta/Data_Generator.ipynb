```python
from passlib.context import CryptContext
from  django.contrib.auth.password_validation import CommonPasswordValidator
import cx_Oracle


#------------------------Encryption Metadata-----------------------------
pwd_context = CryptContext(
        schemes=["pbkdf2_sha256"],
        default="pbkdf2_sha256",
        pbkdf2_sha256__default_rounds=30000
)

#--------------------------Connect Database----------------------------- 
def db():
    dsn_tns = cx_Oracle.makedsn('localhost','1521',service_name='ORCL')
    return cx_Oracle.connect(user='django_db',password='007007',dsn=dsn_tns)



#---------------------------------Password Validation----------------------------
def encrypt_password(password):
    return pwd_context.hash(password)

```

```python
df['Password'] = df['Password'].apply(lambda x :encrypt_password(str(x)))
```

```python
df.to_csv('Database.csv',index=False)
```

```python
conn = db()
c = conn.cursor()
for row in df.iterrows():
    sql = """ INSERT INTO USER_TABLE (STD_ID,FULL_NAME,EMAIL,MOBILE,PASSWORD)
                        VALUES(:std_id,:fullname,:email,:mobile,:password)"""
    try:
        c.execute(sql,{'std_id':row[1][0],'fullname':row[1][1],'email':row[1][4],'mobile':row[1][2],'password':row[1][5]})
        conn.commit()
        print('Registered' + str(row[1][0]))
    except cx_Oracle.IntegrityError:
        message = "User already exists ..."
        print('Error')
    except:
        print('couldn\'t insert '+str(row[1][0]))
conn.close()
```
